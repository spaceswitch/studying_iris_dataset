```python
# speciesの割合の円グラフ
iris['fin_flg'] = iris['species'].map(lambda x: 0 if x=='setosa' else 1 if x=='versicolor' else 2)
# A if 条件式1 else B if 条件式2 else C

iris.groupby('fin_flg').size()
```

```python
species_size=iris.groupby('fin_flg').size()
labels=['setosa','versicolor','virginica']
colors=['red','blue','green']

plt.pie(species_size,labels=labels,colors=colors,startangle = 90,autopct='%1.1f%%')
```

```python
# 特徴行列と目的配列の抽出
X_iris = iris[['sepal_length','sepal_width','petal_length','petal_width']]
X_iris.shape
```

```python
X_iris.head()
```

```python
y_iris = iris['species']
y_iris.shape
```

```python
# ナイーブベイズ分類
X_train,X_test,y_train,y_test = train_test_split(X_iris,y_iris,random_state=1)
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train,y_train)
y_model = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_model)
```

```python
# ロジスティック回帰
from sklearn.linear_model import LogisticRegression
X = iris[['sepal_length','sepal_width','petal_length','petal_width']]
y = iris['fin_flg']
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0,test_size=0.5)
model = LogisticRegression()
model.fit(X_train,y_train)
print('正解率(train):{:.3f}' . format(model.score(X_train,y_train)))
print('正解率(test):{:.3f}' . format(model.score(X_test,y_test)))
```

```python
X = iris[['sepal_length','sepal_width','petal_width']]
y = iris['fin_flg']
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0,test_size=0.5)
model = LogisticRegression()
model.fit(X_train,y_train)
print('正解率(train):{:.3f}' . format(model.score(X_train,y_train)))
print('正解率(test):{:.3f}' . format(model.score(X_test,y_test)))
```

```python
# スケーリングによる予測精度の向上
from sklearn.preprocessing import StandardScaler
X = iris[['sepal_length','sepal_width','petal_length','petal_width']]
y = iris['fin_flg']
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0,test_size=0.5)

sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)

model = LogisticRegression()
model.fit(X_train_std,y_train)
print('正解率(train):{:.3f}' . format(model.score(X_train_std,y_train)))
print('正解率(test):{:.3f}' . format(model.score(X_test_std,y_test)))
```

```python
# サポートベクターマシン
from sklearn.svm import LinearSVC
X = iris[['sepal_length','sepal_width','petal_width']]
y = iris['fin_flg']
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0,test_size=0.5,stratify=y)
model = LinearSVC()
model.fit(X_train,y_train)
print('正解率(train):{:.3f}' . format(model.score(X_train,y_train)))
print('正解率(test):{:.3f}' . format(model.score(X_test,y_test)))
```

```python
# 標準化したサポートベクターマシン
X = iris[['sepal_length','sepal_width','petal_width']]
y = iris['fin_flg']
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0,test_size=0.5,stratify=y)

sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)

model = LinearSVC()
model.fit(X_train_std,y_train)

print('正解率(train):{:.3f}' . format(model.score(X_train_std,y_train)))
print('正解率(test):{:.3f}' . format(model.score(X_test_std,y_test)))
```
